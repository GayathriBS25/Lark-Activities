# Compare churn count with respect to gender

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Telco-Customer-Churn.xlsx")
sns.countplot(df['Churn'],hue=df['gender']).set(title='Comparison between Churn and Gender',ylabel='Count of Male and Female')
```

# Number of Female Citizens

```python

df.columns
#df['SeniorCitizen'].value_counts()
a=df[df.SeniorCitizen==1].groupby('gender').SeniorCitizen.count()
a
```

```python
a.plot(kind='bar',title='Number of Female an Male Citizens',ylabel='Gender Count')
```
