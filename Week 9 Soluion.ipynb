//fill null values with Median

```python
print("Age gets filled with Median in the Nan places")
data['age']=data['age'].fillna(data['age'].median())
data
```

```python
print("Income gets filled with Median in the Nan places")
data['income']=data['income'].fillna(data['income'].median())
data
```

# Check the outliers and handle outliers in performance score using Percentiles

```python
import seaborn as sns
sns.boxplot(data =data, y ='performance_score')
```

```python
Q1=np.percentile(data['performance_score'],25,interpolation='midpoint')
Q2=np.percentile(data['performance_score'],50,interpolation='midpoint')
Q3=np.percentile(data['performance_score'],75,interpolation='midpoint')
m=data['performance_score'].median()
print("Median: ",m)
print(Q1,Q2,Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print("lower_limit: ",lower_limit,"\nupper_limit: ",upper_limit)
```

//Remove Values lesser than lower limit and greater than upper limit

```python
outlier=[]
for x in data['performance_score']:
    if((x>upper_limit) or (x<lower_limit)):
        outlier.append(x)
        
```

# Do the standard scaling on the feature performance score.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

col_names = ["age","income","department","grade","performance_score","gender_M","gender_unknown"]
features = data3[col_names ]
ct = ColumnTransformer([('somename',StandardScaler(),['performance_score'])],remainder ='passthrough')
ct.fit_transform(features)
x_SS = pd.DataFrame(ct.fit_transform(features))

x_SS.columns = ["performance_score","age","income","department","grade","gender_M","gender_unknown"]
x_SS.index = ["Allen Smith","S Kumar","Jack Morgan","Ying Chin","Dheeraj Patel","Satyam Sharma","Josh Wills","Leo Duck"]
x_SS
```
